## Matplotlib example workflow

```python
# 0. Import matplotlib and get it ready for plotting in Jupyter
%matplotlib inline
import matplotlib.pyplot as plt

# 1. Prepare data
x = [1, 2, 3, 4]
y = [11, 22, 33, 44]

# 2. Setup plot
fig, ax = plt.subplots(figsize=(5, 5)) # (width, height)

# 3. Plot data
ax.plot(x, y)

# 4. Customize plot
ax.set(title = 'Simple Plot',
       xlabel = 'x-axis',
       ylabel = 'y-axis')

# 5. Save & show (you save the whole figure)
fig.savefig('images/sample-plot.png')
```

```python
# Make a plot from a dictionary
nut_butter_prices = {'Almond butter': 10,
                     'Peanut butter': 8,
                     'Cashew butter': 12}
fig, ax = plt.subplots(figsize=(7, 3))
ax.bar(nut_butter_prices.keys(), nut_butter_prices.values()) # x, y
ax.set(title="Dan's Nut Butter Store",
       xlabel="Butter Type",
       ylabel="Price ($)");
```

```python
fig, ax = plt.subplots(figsize=(7, 3))
ax.barh(list(nut_butter_prices.keys()), nut_butter_prices.values()) # x, y
ax.set(title="Dan's Nut Butter Store",
       ylabel="Butter Type",
       xlabel="Price ($)");
```

```python
# Make some data for histograms and plot it
x = np.random.randn(1000)

fig, ax = plt.subplots(figsize=(7, 3))
ax.hist(x);
```

### Two options for subplots

```python
# Subplots option 1
fig, ((ax1, ax2), (ax3, ax4)) = plt.subplots(nrows=2, ncols=2, figsize=(10, 5))

# Plot to each different axis
ax1.plot(x, x/2)
ax2.scatter(np.random.random(100), np.random.random(100))
ax3.bar(nut_butter_prices.keys(), nut_butter_prices.values())
ax4.hist(x);
```

```python
# Subplots option 2
fig, ax = plt.subplots(nrows=2, ncols=2, figsize=(10, 5))

# Plot to each different index
ax[0, 0].plot(x, x/2)
ax[0, 1].scatter(np.random.random(100), np.random.random(100))
ax[1, 0].bar(nut_butter_prices.keys(), nut_butter_prices.values())
ax[1, 1].hist(x);
```

### Which one should you use? (pyplot vs matplotlib 00 method?)

* When plotting something quickly, okay to use the pyplot method
* When plotting something more advance, use the OO method (OO = Object Oriented)

```python
heart_disease.head(2)
```

```python
over_50 = heart_disease[heart_disease.age>50]
print(len(over_50))
over_50.head()
```

## Pyplot method

```python
# Pyplot method
over_50.plot(kind='scatter', x='age', y='chol', c='target', figsize=(7, 3));
```

```python
ax = heart_disease.plot(kind='scatter', x='age', y='chol', c='black', figsize=(7, 3));
over_50.plot(kind='scatter', x='age', y='chol', c='red', s=5, ax=ax);
```

## OO method mixed with pyplot method

```python
# OO method mixed with pyplot method
fig, ax = plt.subplots(figsize=(7, 3))

over_50.plot(kind='scatter', x='age', y='chol', c='target', ax=ax);
```

```python
fig, ax = plt.subplots(figsize=(7, 3))

over_50.plot(kind='scatter', x='age', y='chol', c='target', ax=ax)
ax.axhline(y=239, linewidth=0.5, color='red')
ax.set_xlim([45, 100])
ax.legend(['Sample Population over 50', 'Safe limit: 239 mg/dL'], loc='upper right', fontsize=8);
```

```python
fig, ax = plt.subplots(figsize=(7, 3))

over_50.plot(kind='scatter', x='age', y='chol', c='target', label='Sample Population over 50', ax=ax)
ax.axhline(y=239, linewidth=0.5, color='red', label='Safe limit: 239 mg/dL')
ax.set_xlim([45, 100])
ax.legend(loc='upper right', fontsize=8);
```

## OO method from scratch

```python
# OO method from scratch
fig, ax = plt.subplots(figsize=(7, 3))

# Plot the data
scatter = ax.scatter(x=over_50.age, y=over_50.chol, c=over_50.target);

# Customize the plot
ax.set(title='Heart Disease and Cholesterol Levels',
       xlabel='Age',
       ylabel='Cholesterol')

# Add a legend
print(over_50.target.values)
print(scatter.legend_elements())
ax.legend(*scatter.legend_elements(), title='Target');
```

```python
# OO method from scratch
fig, ax = plt.subplots(figsize=(7, 3))

# Plot the data
scatter = ax.scatter(x=over_50.age, y=over_50.chol, c=over_50.target);

# Customize the plot
ax.set(title='Heart Disease and Cholesterol Levels',
       xlabel='Age',
       ylabel='Cholesterol')

# Add a legend
ax.legend(*scatter.legend_elements(), title='Target')

# Add a horizontal line
ax.axhline(over_50.chol.mean(), linestyle='--');
```

## Subplots

```python
# Subplot of chol, age, thalach
fig, (ax0, ax1) = plt.subplots(nrows=2, ncols=1, figsize=(7, 6))
fig.suptitle('Heart Disease', color='darkblue')

# Add data to ax0
scatter = ax0.scatter(x=over_50.age, y=over_50.chol, c=over_50.target)

# Customize ax0
ax0.set(title='Cholesterol Levels',
       xlabel='Age',
       ylabel='Cholesterol')

# Add legend to ax0
ax0.legend(*scatter.legend_elements(), title='Target')

# Add a meanline
ax0.axhline(over_50.chol.mean(), linestyle='--')

# Add data to ax1
scatter = ax1.scatter(x=over_50.age, y=over_50.thalach, c=over_50.target)

# Customize ax1
ax1.set(title='Max Heart Rate',
        xlabel='Age',
        ylabel='Max Heart Rate')

# Add legend to ax1
ax1.legend(*scatter.legend_elements(), title='Target')

# Add a meanline to ax1
ax1.axhline(over_50.thalach.mean(), linestyle='--')

plt.tight_layout();
```

#### Sharing x axis

```python
# Subplot of chol, age, thalach
fig, (ax0, ax1) = plt.subplots(nrows=2, ncols=1, figsize=(7, 6), sharex=True)

# Add a title to the figure
fig.suptitle('Heart Disease', color='darkblue', fontsize=16, fontweight='bold')

# Add data to ax0
scatter = ax0.scatter(x=over_50.age, y=over_50.chol, c=over_50.target)

# Customize ax0
ax0.set(title='Cholesterol Levels',
        ylabel='Cholesterol')

# Add legend to ax0
ax0.legend(*scatter.legend_elements(), title='Target')

# Add a meanline
ax0.axhline(over_50.chol.mean(), linestyle='--')

# Add data to ax1
scatter = ax1.scatter(x=over_50.age, y=over_50.thalach, c=over_50.target)

# Customize ax1
ax1.set(title='Max Heart Rate',
        xlabel='Age',
        ylabel='Max Heart Rate')

# Add legend to ax1
ax1.legend(*scatter.legend_elements(), title='Target')

# Add a meanline to ax1
ax1.axhline(over_50.thalach.mean(), linestyle='--')

plt.tight_layout();
```

## Customizing the y and x axis limitations

![](../zero-to-mastery-ml/images/matplotlib-anatomy-of-a-plot.png)

This plots shows some information about the heart disease dataset...

```python
# Subplot of chol, age, thalach
fig, (ax0, ax1) = plt.subplots(nrows=2, ncols=1, figsize=(7, 6), sharex=True)

# Add a title to the figure
fig.suptitle('Heart Disease Analysis', color='darkblue', fontsize=16, fontweight='bold')

# Add data to ax0
scatter = ax0.scatter(x=over_50.age, y=over_50.chol, c=over_50.target, cmap='winter')

# Customize ax0
ax0.set(title='Heart Disease and Cholesterol Levels',
        ylabel='Cholesterol')
ax0.set_xlim([50, 80]) # Change the x axis limits

# Add legend to ax0
ax0.legend(*scatter.legend_elements(), title='Target')

# Add a meanline
ax0.axhline(over_50.chol.mean(), linestyle='--')

# Add data to ax1
scatter = ax1.scatter(x=over_50.age, y=over_50.thalach, c=over_50.target, cmap='winter')

# Customize ax1
ax1.set(title='Heart Disease and Max Heart Rate',
        xlabel='Age',
        ylabel='Max Heart Rate')

ax1.set_ylim([60, 200]) # Change the y axis limits

# Add legend to ax1
ax1.legend(*scatter.legend_elements(), title='Target')

# Add a meanline to ax1
ax1.axhline(over_50.thalach.mean(), linestyle='--')

plt.tight_layout();
```
